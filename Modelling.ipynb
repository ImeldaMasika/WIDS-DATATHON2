###### Working on the breast_cancer_diagnosis_desc column

```python
# Dictionary mapping descriptions to categories
diagnosis_categories = {
    'Upper Quadrants': ['upper-outer quadrant', 'upper-inner quadrant'],
    'Lower Quadrants': ['lower-outer quadrant', 'lower-inner quadrant'],
    'Central Portion': ['central portion'],
    'Nipple and Areola': ['nipple and areola'],
    'Axillary Tail': ['axillary tail'],
    'Overlapping Sites': ['ovrlp sites', 'overlapping sites'],
    'Unspecified Sites': ['unspecified site', 'unsp site']
}

# Function to map description to category
def map_diagnosis_to_category(description):
    for category, keywords in diagnosis_categories.items():
        for keyword in keywords:
            if keyword in description.lower():
                return category
    return 'Other'  # If no match is found, categorize as 'Other'

# Apply the mapping function to create a new column 'diagnosis_category'
new_train['diagnosis_category'] = new_train['breast_cancer_diagnosis_desc'].apply(map_diagnosis_to_category)
```

#### Marital Status

```python
# Dictionary mapping each marital status category to its corresponding point value
marital_points = {
    'married': 1,
    'divorced': 2,
    'never_married': 3,
    'widowed': 4
}


# Calculate the weighted average marital status for each row
new_train['marital_status'] = (
    new_train['married'] * marital_points['married'] +
    new_train['divorced'] * marital_points['divorced'] +
    new_train['never_married'] * marital_points['never_married'] +
    new_train['widowed'] * marital_points['widowed']
) / 100

# Apply ordinal encoding to represent marital status numerically
new_train['marital_status_encoded'] = pd.cut(new_train['marital_status'], bins=4, labels=False) + 1
```

#### Chi- Square and ANOVA

```python
from scipy.stats import chi2_contingency
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Chi-square test for region and division
cross_tab = pd.crosstab(new_train['Region'], new_train['Division'])
chi2, p, dof, expected = chi2_contingency(cross_tab)
print("Chi-square test results:")
print("Chi-square:", chi2)
print("p-value:", p)
```

Since the p-value is very close to zero, it is below any reasonable significance level (e.g., 0.05), indicating strong evidence against the null hypothesis. Therefore, I will reject the null hypothesis of no association between the variables. This suggests that there is a significant association between the region and division variables in the dataset.

```python
# ANOVA for patient_zip3, region, and division
model = ols('patient_zip3 ~ C(Region) + C(Division)', data=new_train).fit()
anova_results = anova_lm(model)
print("\nANOVA results:")
print(anova_results)
```

Both Region and Division have a significant impact on patient_zip3, suggesting that there are statistically significant differences in patient_zip3 among the different regions and divisions.

I will drop Region and Division and work with patient_zip3

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
X = new_train.drop(columns=['metastatic_diagnosis_period'])  
y = new_train['metastatic_diagnosis_period']  
```

```python
# Initialize SelectKBest with chi2
k = 35  # Number of top features to select
selector = SelectKBest(score_func=chi2, k=k)

# Fit selector to features and transform the data
X_selected = selector.fit_transform(X, y)

# Get the indices of the selected features
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features = X.columns[selected_indices]

# Print the selected feature names
print(selected_features)
```

```python
# Create a new DataFrame with only the selected features
df_selected = new_train[selected_features]

# Add the target variable (metastatic_diagnosis_period) to the selected features DataFrame
df_selected['metastatic_diagnosis_period'] = new_train['metastatic_diagnosis_period']

# Print the new DataFrame
df_selected.head()
```

#### Modelling

##### LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVR
```

```python
# Split the data into features (X) and target variable (y)
X = df_selected.drop(columns=['metastatic_diagnosis_period'])  
y = df_selected['metastatic_diagnosis_period']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Instantiate the linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X_train, y_train)
```

```python
# Make predictions on the training data
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r_squared = r2_score(y_test, predictions)

print("Mean Squared Error:", mse)
print("R-squared:", r_squared)
```

### SVR Model

```python
# Initialize the SVR model 
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
# train the model on the training data
svr_model.fit(X_train, y_train)

# Then Make predictions on the training data
y_pred_train_svr = svr_model.predict(X_train)

# Make predictions on the testing data
y_pred_test_svr = svr_model.predict(X_test)
```

```python
# Evaluate the model
train_mse_svr = mean_squared_error(y_train, y_pred_train_svr)
test_mse_svr = mean_squared_error(y_test, y_pred_test_svr)
train_r2_svr = r2_score(y_train, y_pred_train_svr)
test_r2_svr = r2_score(y_test, y_pred_test_svr)

print(f"SVR - Train MSE: {train_mse_svr}, Test MSE: {test_mse_svr}")
print(f"SVR - Train R2: {train_r2_svr}, Test R2: {test_r2_svr}")
```
